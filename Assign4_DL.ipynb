```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data = pd.read_csv("salary.csv")
data.head()
```

```python
X = data[["age", "experience"]].values
y = data["income"].values
```

```python
n_features = X.shape[1]
w = np.zeros(n_features)
b = 0.0
```

```python
def predict(X, w, b):
    return X.dot(w) + b
def mean_squared_error(y, y_hat):
    loss = ((y_hat - y) ** 2).mean()
    return loss
```

```python
def compute_gradients(X, y, y_hat):
    N = len(y)
    dw = (2/N)*X.T.dot(y_hat - y)
    db = (2/N)*(y_hat - y).sum()
    return dw,db

def update_parameters(w, b, dw, db, lr):
    w = w - lr*dw
    b = b - lr*db
    return w,b
```

```python
lr =0.00001
epochs =1000
for epoch in range(epochs):
    y_hat = predict(X, w, b)
    current_loss = mean_squared_error(y, y_hat)
    dw,db = compute_gradients(X, y, y_hat)
    w,b = update_parameters(w, b, dw, db, lr)
    if epoch % 100 == 0:
        print(f"Epoch {epoch},Loss:{current_loss}")
```

```python
new_candidate = np.array([35, 8])
predicted_income = new_candidate.dot(w)+b

print(f"Final Weights: {w}")
print(f"Final Bias: {b}")
print(f"Predicted Income: {predicted_income}")
```
